## Popular books

```python
#merging rating and books with ISBN
ratings_w_name= ratings.merge(books,on='ISBN')
ratings_w_name
```

```python
#counting the number of ratings each book has
num_rating=ratings_w_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_rating.rename(columns={'Book-Rating':'num_ratings'},inplace=True)
num_rating 
```

```python
#average of all the ratings each book has
avg_rating=ratings_w_name.groupby('Book-Title').mean()['Book-Rating'].reset_index()
avg_rating.rename(columns={'Book-Rating':'avg_rating'},inplace=True)
avg_rating
```

```python
#merging number of ratings and average ratings
popular=num_rating.merge(avg_rating, on='Book-Title')
popular
```

```python
#Considering only those books which have  more than 220 book ratings and sorting them descending
popular=popular[popular['num_ratings']>=220].sort_values('avg_rating',ascending=False).head(10)
popular
```

```python

popular_df=popular.merge(books,on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author','Image-URL-M','num_ratings','avg_rating']]
```

```python
popular_df
```

```python
#importing pickle library and dumping the file in our directory
import pickle
pickle.dump(popular_df,open('popular_df.pkl','wb'))
```

```python

ratings_w_name= ratings.merge(books,on='ISBN')
ratings_w_name
```

```python
#Number of ratings each user has rated
users_with_ratings=ratings_w_name.groupby("User-ID").count()['Book-Rating']
users_with_ratings
```

```python
#users who have rated more than 200
x=  ratings_w_name.groupby("User-ID").count()['Book-Rating']>200
x
```

```python
x[x]
```

```python
#Retreiving the user-id's of the users who have rated more than 200 books
users_with_more_ratings= x[x].index
users_with_more_ratings
```

```python
#filtering the dataset with users with more ratings
filtered_ratings= ratings_w_name[ratings_w_name['User-ID'].isin(users_with_more_ratings)]
filtered_ratings
```

```python
#grouping by book title and book-rating
filtered_ratings.groupby('Book-Title').count()['Book-Rating']
```

```python
#considering those books which have more than 50 ratings
y=filtered_ratings.groupby('Book-Title').count()['Book-Rating']>50
y

```

```python
#retreiving them
famous_books=y[y].index
famous_books
```

```python
final_ratings=filtered_ratings[filtered_ratings['Book-Title'].isin(famous_books)]
```

```python
final_ratings
```

```python
#dropping duplicates if any
final_ratings.drop_duplicates()
```

## Recommend function

```python
#finding a particular book
import numpy as np

np.where(ptable.index=='Year of Wonders')[0][0]
    
```

```python
#using the lambda function 
similar_books=sorted(list(enumerate(sim_scores[0])),key=lambda x:x[1],reverse=True)[1:9]
similar_books
```

```python
#the recommend function
def recommend(book_name):
    index=np.where(ptable.index==book_name)[0][0]
    similar_books=sorted(list(enumerate(sim_scores[index])),key=lambda x:x[1],reverse=True)[1:9]
    data=[]
    for i in similar_books:
        item = []
        temp_df = books[books['Book-Title'] == ptable.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))
        
        data.append(item)
    
    return data
    
    
```

## Recommending books
